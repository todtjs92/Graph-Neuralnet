## Setup

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

from torch_geometric.nn import GCNConv

import torch_geometric.transforms as T
from ogb.nodeproppred import PygNodePropPredDataset, Evaluator
```

## Load and Preprocess the Dataset

```python
dataset_name = 'ogbn-arxiv'
dataset = PygNodePropPredDataset(name=dataset_name,
                                 transform=T.ToSparseTensor())
data = dataset[0]

# Make the adjacency matrix to symmetric
data.adj_t = data.adj_t.to_symmetric()

device = 'cuda' if torch.cuda.is_available() else 'cpu'

# If you use GPU, the device should be cuda
print('Device: {}'.format(device))

data = data.to(device)
split_idx = dataset.get_idx_split()
train_idx = split_idx['train'].to(device)
```

## 실습 2: GCN Model

![test](https://drive.google.com/uc?id=128AuYAXNXGg7PIhJJ7e420DoPWKb-RtL)
- [팁] 훈련중에만 dropout 적용하기
    - 아래의 코드를 모델 클래스의 forward 함수 내부에 사용해줍니다.

    `F.dropout(x, p=self.dropout, training=self.training)`

- [팁] torch.nn 모듈을 리스트 형태로 넣기
    
    `self.linears = nn.ModuleList([nn.Linear(10, 10) for i in range(10)])`
    - Document: https://pytorch.org/docs/stable/generated/torch.nn.ModuleList.html

- [주의사항] 미리 정의된 input, output은 바꾸지 않습니다.
- [주의사항] 모델 코드는 입력된 `num_layers`의 갯수에 따라 모델 레이어를 구성하도록 정의되어야 합니다. (만약 num_layers가 10이면, "GCNConv-BN-ReLU-Dropout"이 9번 반복되고, 마지막 레이어에서 "GCNConv-LogSoftmax"가 정의되어야 합니다.)

```python
class GCN(torch.nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, num_layers,
                 dropout, return_embeds=False):
        super(GCN, self).__init__()

        # A list of GCNConv layers
        self.convs = None

        # A list of 1D batch normalization layers
        # nn.BatchNorm1d(out)
        self.bns = None

        self.softmax = None
        self.num_layers = num_layers

        ############# Your code here ############
       
        self.convs = nn.ModuleList([GCNConv(input_dim, hidden_dim)])
        self.convs.extend([GCNConv(hidden_dim, hidden_dim) for i in range(num_layers-2)])
        self.convs.extend([GCNConv(hidden_dim, output_dim)])
        
        self.bns = nn.ModuleList([nn.BatchNorm1d(hidden_dim) for i in range(num_layers-1)])
        
        self.softmax = nn.LogSoftmax(dim=1)
        #########################################

        # Probability of an element to be zeroed
        self.dropout = dropout

        # Skip classification layer and return node embeddings
        self.return_embeds = return_embeds

    def reset_parameters(self):
        for conv in self.convs:
            conv.reset_parameters()
        for bn in self.bns:
            bn.reset_parameters()

    def forward(self, x, adj_t):
        out = None

        ############# Your code here ############
        for i in range(self.num_layers):
            
            # Last Conv layer pass
            if (i == self.num_layers-1):
                x = self.convs[i](x, adj_t)
                if (self.return_embeds):
                    return x
                x = self.softmax(x)
                out = x
                
            else:
                x = self.convs[i](x, adj_t)
                x = self.bns[i](x)
                x = F.relu(x)
                x = F.dropout(x, p=self.dropout, training=self.training)
        #########################################

        return out
```

## 실습 3: Training GCN Model
- `train(model, data, train_idx, optimizer, loss_fn)` 모델 훈련이 1번 진행되는 함수입니다. input을 이용해 모델의 계산값인 output을 구하고, 해당 output 중에서 `train_index`에 속하는 노드들만들 선택해서, 실제 label과 예측 label의 loss값을 구하세요.
- [주의사항] 미리 정의된 input, output은 바꾸지 않습니다.

```python
def train(model, data, train_idx, optimizer, loss_fn):
    model.train()
    loss = 0

    ############# Your code here ############
    optimizer.zero_grad()
    out = model(data.x, data.adj_t)
    loss = loss_fn(out[train_idx], data.y[train_idx].squeeze(1))
    #########################################

    loss.backward()
    optimizer.step()

    return loss.item()
```

## 실습 4: Evaluation loop

```python
@torch.no_grad() # for evaluation mode
def test(model, data, split_idx, evaluator):
    model.eval() # for evaluation mode

    out = None

    ############# Your code here ############
    out = model(data.x, data.adj_t)
    #########################################

    y_pred = out.argmax(dim=-1, keepdim=True)

    train_acc = evaluator.eval({
        'y_true': data.y[split_idx['train']],
        'y_pred': y_pred[split_idx['train']],
    })['acc']
    valid_acc = evaluator.eval({
        'y_true': data.y[split_idx['valid']],
        'y_pred': y_pred[split_idx['valid']],
    })['acc']
    test_acc = evaluator.eval({
        'y_true': data.y[split_idx['test']],
        'y_pred': y_pred[split_idx['test']],
    })['acc']

    return train_acc, valid_acc, test_acc
```

## Experiments

```python
# Please do not change the args
args = {
    'device': device,
    'num_layers': 3,
    'hidden_dim': 256,
    'dropout': 0.5,
    'lr': 0.01,
    'epochs': 100,
}
args
```

```python
model = GCN(data.num_features, args['hidden_dim'],
            dataset.num_classes, args['num_layers'],
            args['dropout']).to(device)
evaluator = Evaluator(name='ogbn-arxiv')
```

```python
import copy
from tqdm.notebook import tqdm

# reset the parameters to initial random value
model.reset_parameters()

optimizer = torch.optim.Adam(model.parameters(), lr=args['lr'])
loss_fn = F.nll_loss

best_model = None
best_valid_acc = 0
pbar = tqdm(range(1, 1 + args["epochs"]))

for epoch in range(1, 1 + args["epochs"]):
  loss = train(model, data, train_idx, optimizer, loss_fn)
  result = test(model, data, split_idx, evaluator)
  train_acc, valid_acc, test_acc = result
  if valid_acc > best_valid_acc:
      best_valid_acc = valid_acc
      best_model = copy.deepcopy(model)
  pbar.set_description(f'Epoch: {epoch:02d}, Loss: {loss:.4f}, Train: {100 * train_acc:.2f}% Valid: {100 * valid_acc:.2f}% Test: {100 * test_acc:.2f}%')
  pbar.update()

pbar.close()
```

```python
best_result = test(best_model, data, split_idx, evaluator)
train_acc, valid_acc, test_acc = best_result
print(f'Best model: '
      f'Train: {100 * train_acc:.2f}%, '
      f'Valid: {100 * valid_acc:.2f}% '
      f'Test: {100 * test_acc:.2f}%')
```
